#### Calculate Profit
Suppose that you are a CEO of Apple and you receive the following two figures of data from your CFO that shows
the revenue and expenses of the last three years as shown in the figure below. Now you would like to calculate the profit/loss as in table 3.





```python
revenue  = np.array([[180,200,220],[24,36,40],[12,18,20]])
expenses = np.array([[80,90,100],[10,16,20],[8,10,10]])
```

```python
profit = revenue -  expenses
profit
```
